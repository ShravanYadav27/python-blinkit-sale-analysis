```python
Total_Sale = df['Sales'].sum()
```

```python
avg_sales = df['Sales'].mean()
```

```python
print("Total Sales:", Total_Sale)
print("Average Sales:", avg_sales)
print("No of items sold:", no_of_items_sold)
print("Average Rating:",avg_rating)
```

```python
#Q.Total Sales by Fat Content

Sales_by_Fat=df.groupby("Item Fat Content")["Sales"].sum()
print(Sales_by_Fat)
```

```python
plt.pie(Sales_by_Fat, labels=Sales_by_Fat, autopct='%.1f%%', startangle=90)
plt.title('Sales By Fat Content')
plt.axis('equal')

plt.show()
```

```python
#Q. Total Sales by Item Type
```

```python
total_sales = df.groupby("Item Type")["Sales"].sum().sort_values(ascending=False)
print(total_sales)

plt.figure(figsize=(12, 6))  # Optional: Better size
plt.bar(total_sales.index, total_sales.values, color='skyblue')

plt.title("Total Sales by Item Type")
plt.xlabel("Item Type")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

plt.show()
```

```python
#Q.Fat Content by outlet for Total Sales
```

```python
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Sales'].sum().unstack()

grouped.plot(kind='bar', figsize=(12, 6))

plt.title("Total Sales by Fat Content and Outlet")
plt.xlabel("Outlet")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
plt.legend(title='Fat Content')
plt.tight_layout()

plt.show()
```

```python
#Q.Total Sales By Outlet Establishment Year
```

```python
Sales_Estab = df.groupby("Outlet Establishment Year")["Sales"].sum()

plt.figure(figsize=(12, 6))
plt.plot(Sales_Estab.index, Sales_Estab.values, color="blue", marker='o', label="Total Sales")

plt.title("Total Sales By Outlet Establishment Year")
plt.xlabel("Outlet Establishment Year")
plt.ylabel("Total Sales")
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
#Q.Sales By Outlet Size	
```

```python
Sales_Out_size = df.groupby("Outlet Size")["Sales"].sum()

plt.figure(figsize=(8, 5))
plt.pie(Sales_Out_size.values, labels=Sales_Out_size.index,autopct='%1.f%%',startangle=90,colors=['Red','Blue','Green'])

plt.title("Total Sales by Outlet Size")
plt.axis('equal')
plt.tight_layout()

plt.show()
```

```python
#Q.Sales By Outlet Location 
```

```python
Out_Locn = df.groupby("Outlet Location Type")["Sales"].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 6))
plt.barh(Out_Locn.index, Out_Locn.values, color='orange')

plt.title("Sales Funnel by Outlet Location Type")
plt.xlabel("Total Sales")
plt.ylabel("Outlet Location Type")
plt.gca().invert_yaxis()
plt.tight_layout()

plt.show()
```
